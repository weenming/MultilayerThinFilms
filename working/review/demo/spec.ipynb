```python
import sys
sys.path.append('./../../../')
sys.path.append('./../../../designer/script/')
sys.path.append('./../../')


import numpy as np
import matplotlib.pyplot as plt

from film import TwoMaterialFilm

wls = np.linspace(200, 900, 500)
def make_bragg(d):
    modulated_bragg = TwoMaterialFilm('1', '2', 'SiO2', d, incidence='Air')
    modulated_bragg.add_spec_param(0., wls)
    return modulated_bragg
l = 100
d = np.array([l, 0.5 * l] * 4)

fig, ax = plt.subplots(1, 1)
ax.plot(wls, make_bragg(d).get_spec().get_R(), label=f'{l}', linewidth=6)

fig.set_size_inches(5, 3)
fig.savefig('./demo_spectrum.png', dpi=300)
```
