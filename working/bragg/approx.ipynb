```python
# approx: Approximate calculation of the reflection coefficient from a stratified medium
# Sheppard, C. J.R.
import numpy as np
import matplotlib.pyplot as plt

X, Y = np.meshgrid(np.linspace(-2 * np.pi, 2 * np.pi, 100), np.linspace(-2 * np.pi, 2 * np.pi, 100))
Z = 2 * np.sin((X + Y) / 2) * np.sin((X - Y) / 2)
Z_1 = np.zeros_like(X) + 1
fig, ax = plt.subplots(1, 1, subplot_kw=dict(projection='3d'))
ax.plot_surface(X, Y, np.abs(Z))
# ax.plot_surface(X, Y, Z_1)
```
