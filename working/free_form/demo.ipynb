```python
wls = np.linspace(500, 700, 500)

n_size = 100
each_ot = 100 # nm
init_n = np.array([1.5, 2.5] * 20)

film = EqOTFilm(init_n, each_ot * n_size, substrate=1.5)



rs = film.get_spec(60., wls).get_R(s_ratio=0).copy()
rp = film.get_spec(60., wls).get_R(p_ratio=0).copy()
ts = film.get_spec(60., wls).get_T(s_ratio=0).copy()
tp = film.get_spec(60., wls).get_T(p_ratio=0).copy()
r = film.get_spec(60., wls).get_R().copy()


fig, ax = plt.subplots(1, 1)
ax.plot(wls, rs / ts, label='s')
ax.plot(wls, rp / tp, label='p')
ax.legend()
fig.set_size_inches(3, 2)
ax.set_xlabel('Wavelength / nm')
ax.set_ylabel('Electric field')


fig, ax = plt.subplots(1, 1)
ax.plot(wls, rs / ts / (rp / tp))
fig.set_size_inches(3, 2)
ax.set_xlabel('Wavelength / nm')
ax.set_ylabel('$\\tan^2 \\Psi$')
```
