```python
import numpy as np
import matplotlib.pyplot as plt

xmin = 0.1
xmax = 5
X, Y = np.meshgrid(np.linspace(xmin, xmax, 100), np.linspace(xmin, xmax, 100))
L = - 1.8 * Y * X + X ** 2 + Y ** 2

print(X.shape)
fig, ax = plt.subplots(1, 1)
s = ax.contourf(
    np.log10(L), 
    cmap='viridis', 
    extent=[xmin, xmax, xmin, xmax]
)
cbar = fig.colorbar(s)
cbar.set_label('log10 loss')

n_c_min = 1.5
n_c_max = 3
ax.plot([n_c_min, n_c_max], [n_c_min, n_c_min], label='constraint', c='C0')
ax.plot([n_c_min, n_c_max], [n_c_max, n_c_max], c='C0')
ax.plot([n_c_min, n_c_min], [n_c_min, n_c_max], c='C0')
ax.plot([n_c_max, n_c_max], [n_c_min, n_c_max], c='C0')
```
